```python
import numpy as np
import pandas as pd
import requests
```

```python
Url = "https://cea.nic.in/api/electricEnergySales.php"
```

```python
response = requests.get(Url)
```

```python
if response.status_code == 200:
  data = response.json()
  df = pd.DataFrame(data)
  print("Data is ready...")
else:
  print(f"Failed to fetch the data. Status code: {response.status_code}")
```
